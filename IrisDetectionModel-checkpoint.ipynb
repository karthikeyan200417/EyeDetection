```python
!pip install numpy opencv-python-headless scikit-image matplotlib scikit-learn joblib

# Import required libraries
import os
import cv2
import numpy as np
from skimage.feature import hog
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
import joblib
```

```python
# Load the pre-trained Haar cascade for eye detection
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
# Function to extract HOG features from an image
def extract_hog_features(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Resize image to a fixed size (optional, adjust as needed)
    resized_image = cv2.resize(gray, (64, 64))
    
    # Compute HOG features
    hog_features = hog(resized_image, orientations=9, pixels_per_cell=(8, 8),
                       cells_per_block=(2, 2), block_norm='L2-Hys', visualize=False, transform_sqrt=True)
    
    return hog_features
```

```python
# Function to segment eye image, extract features, and visualize
def segment_and_extract_features(image_path,sample):
    # Load the image
    image = cv2.imread(image_path)
    
    # Convert image to grayscale for Haar cascade
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Detect eyes in the image
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
    
    # Initialize list to store extracted features
    eye_features = []
    
    # Iterate over detected eyes
    for i, (ex, ey, ew, eh) in enumerate(eyes):
        # Extract the eye region from the image
        eye_region = image[ey:ey+eh, ex:ex+ew]
        
        # Extract HOG features from the eye region
        hog_features = extract_hog_features(eye_region)
        
        # Store the eye region and its features
        eye_features.append((eye_region, hog_features))
    if sample == 1:
        # Display segmented eye images and their HOG features
        fig, axs = plt.subplots(1, len(eye_features), figsize=(12, 4))

        for i, (eye_image, hog_features) in enumerate(eye_features):
            axs[i].imshow(cv2.cvtColor(eye_image, cv2.COLOR_BGR2RGB))
            axs[i].set_title(f'Eye {i+1}')
            axs[i].axis('off')

            # Print HOG features for the eye
            print(f"Extracted HOG Features for Eye {i+1}:")
            print(hog_features)

        plt.tight_layout()
        plt.show()
        
    return eye_features if len(eye_features) > 0 else None
```

```python
# Function to load eye and non-eye images and extract features
def load_eye_and_non_eye_features(data_dir):
    eye_features = []
    non_eye_features = []
    
    # Load eye images and extract features
    eye_dir = os.path.join(data_dir, 'eye')
    for filename in os.listdir(eye_dir):
        if filename.endswith('.jpeg') or filename.endswith('.png'):
            image_path = os.path.join(eye_dir, filename)
            features = segment_and_extract_features(image_path,0)
            if features:
                eye_features.extend(features)
    
    # Load non-eye images and extract features
    non_eye_dir = os.path.join(data_dir, 'non-eye')
    for filename in os.listdir(non_eye_dir):
        if filename.endswith('.jpeg') or filename.endswith('.png'):
            image_path = os.path.join(non_eye_dir, filename)
            image = cv2.imread(image_path)
            hog_features = extract_hog_features(image)
            non_eye_features.append(hog_features)
    
    # Convert lists to numpy arrays
    eye_features = np.array([feat[1] for feat in eye_features])
    non_eye_features = np.array(non_eye_features)
    
    # Create labels (1 for eye, 0 for non-eye)
    eye_labels = np.ones(len(eye_features))
    non_eye_labels = np.zeros(len(non_eye_features))
    
    return eye_features, non_eye_features, eye_labels, non_eye_labels
```

```python
data_dir = 'data'  # Replace with your data directory
```

```python
# Load and process an example image from the eye folder
eye_image_files = os.listdir(os.path.join(data_dir, 'eye'))
eye_image_path = os.path.join(data_dir, 'eye', eye_image_files[2])
eye_features = segment_and_extract_features(eye_image_path,1)
```

```python
# Load eye and non-eye features and labels
eye_features, non_eye_features, eye_labels, non_eye_labels = load_eye_and_non_eye_features(data_dir)
```

```python
# Combine features and labels
X = np.vstack((eye_features, non_eye_features))
y = np.hstack((eye_labels, non_eye_labels))
```

```python
# Perform PCA to reduce dimensionality to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
# Train SVM classifier
svm_clf = SVC(kernel='linear', C=1.0, random_state=42)
svm_clf.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = svm_clf.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM classifier: {accuracy}")
```

```python
# Plotting decision boundary
plt.figure(figsize=(10, 6))

# Plot data points with color labels
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', edgecolor='k', s=100)
plt.title('PCA of Eye and Non-Eye Images')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

# Add legend for colors
plt.legend(handles=scatter.legend_elements()[0], labels=['Non-Eye', 'Eye'], loc='upper right')

# Plotting decision regions
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svm_clf.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
ax.scatter(svm_clf.support_vectors_[:, 0], svm_clf.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```

```python
def predict_eye(image_path, svm_classifier, pca_transformer):
    # Load the image
    image = cv2.imread(image_path)
    
    # Check if the image is loaded properly
    if image is None:
        print("Error: Could not load image.")
        return
    
    # Extract HOG features from the image
    features = segment_and_extract_features(image_path, sample=0)
    
    if not features:
        print("No eyes detected in the image.")
        return
    
    # Extract the HOG features from the detected eye regions
    hog_features = np.array([feat[1] for feat in features])
    
    # Apply PCA transformation
    hog_features_pca = pca_transformer.transform(hog_features)
    
    # Predict using the SVM classifier
    predictions = svm_classifier.predict(hog_features_pca)
    
    # Print and return the prediction results
    for i, prediction in enumerate(predictions):
        if prediction == 1:
            print(f"Eye {i + 1}: Detected as an Eye")
        else:
            print(f"Eye {i + 1}: Detected as Non-Eye")
    
    return predictions

# Example usage:
example_image_path = '5.jpeg'  # Replace with your image path
predict_eye(example_image_path, svm_clf, pca)
```
